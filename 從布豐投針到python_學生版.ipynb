###蒙地卡羅法
####一種數值方法，利用亂數取樣(random sampling)模擬來解決數學問題。
* 我們用它來計算圓周率

image.png


```python
from PIL import Image
from math import *
import random
size_sample=1000000
m=1000;n=1000                  
new=Image.new('RGB',(m,n),(255,255,255))
numb=0
for i in range(size_sample):
    d=[random.random(),random.random()]
    h=int(1000*d[0])
    k=int(1000*d[1])

    if (d[0]-0.5)**2+(d[1]-0.5)**2<=0.25:
      numb=numb+1
      new.putpixel((h,k),(0,0,255))
    else:
      new.putpixel((h,k),(0,255,0))
print("pi",4*numb/size_sample)
new.save('Monte Carlo method.png')
#print('Done')
```

#####1.丟硬幣正面機率是多少?
* 重複丟擲硬幣
* 計數正面出現次數
* 計算比值

##我們常利用計算機幫忙做
* 重複性高的(規律的)
* 數量大的


```python
from math import *
import random
n = int(input("請輸入實驗次數:"))
count = 0
for i in range(n):
    d=random.random()
    mod = int(10*d) % 2
    if mod == 0:
      count=count+1
print("出現正面的機率為",count/n)
 
```

1. 優 S$\geq $90
2. 甲 90$>$S$\geq $80

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import *
myList = input("請輸入國、英、數的成績:").split()
score = [int(string) for string in myList ]
print(score)
T=sum(score)
print("總分:",T)
s=T/3
if s>=90:
 print("%s %3.1f %s" %('平均為',s,'等第:優'))
elif 90>s>=80:
  print("%s %3.1f %s" %('平均為',s,'等第:甲'))
elif 80>s>=70:
  print("%s %3.1f %s" %('平均為',s,'等第:乙'))
elif 70>s>=60:
  print("%s %3.1f %s" %('平均為',s,'等第:丙'))
elif 60>s>=50:
  print("%s %3.1f %s" %('平均為',s,'等第:丁'))
else:
  print("%s %3.1f %s" %('平均為',s,'等第:戊'))
```

#####2.輸入三數 $a,b,c，$判斷是否為直角三角形的邊長。

######提示: $a^2 + b^2 = c^2$

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import *
myList = input("請輸入a、b、c:").split() 
A = [int(string) for string in myList ]
A.sort()
print(A)
a=A[0]
b=A[1]
c=A[2]
print(A)
if a**2+b**2 == c**2:
  print("%d %d %d %s" %(a,b,c,"為直角三角形的邊長"))
else:
  print("%d %d %d %s" %(a,b,c,"不為直角三角形的邊長")) 
```

#####1.費氏數列
* 1、1、2、3、5、8、13、21、34…..為費氏數列，其前兩為1、1，其餘各項皆為其前兩項之和，輸出費氏數列的前31項。
* $\begin{cases}a_{1}=1\\ a_{2}=1\\ a_{n+2}=a_{n+1}+a_{n},n\geq 1\end{cases} $

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import *
n=int(input("請數入項數"))
if  n==1:
 print("費氏數列的第%d項為" %(n,1))
elif n==2:
 print("費氏數列的第%d項為" %(n,1)) 
else:
  a = 1
  b = 1  
  print(a,b)
  for i in range(n-2):
    k = a + b
    a=b
    b=k
    print(k)
  print("費氏數列的第%d項為 %d" %(n,k))
```
